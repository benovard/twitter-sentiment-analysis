```python
"""Plots data in an interactive matplotlib plot"""
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
from ipywidgets import widgets, interactive
```

```python
df = pd.read_csv('../tweet_data/matrix_data.csv', header=0)
```

```python
select1 = widgets.Dropdown(
    options=list(df.columns),
    description='first keyword'
)

select2 = widgets.Dropdown(
    options=list(df.columns),
    description='second keyword'
)
```

```python
import matplotlib.dates as mdates
myFmt = mdates.DateFormatter('%d')
```

```python
def plot(select1, select2):

    ax = plt.gca()
    #ax.xaxis_date()
    #ax.xaxis.set_major_formatter(myFmt)
    df.plot(kind='line', x='date', y=select1, ax=ax)
    df.plot(kind='line', x='date', y=select2, ax=ax)
    plt.axhline(color='black')
    plt.ylim((-1, 1))
    plt.show()
```

```python
interactive(plot, select1=select1, select2=select2)
```
